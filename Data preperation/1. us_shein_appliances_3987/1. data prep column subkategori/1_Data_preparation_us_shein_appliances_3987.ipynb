```python
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text

vectorizer = CountVectorizer(ngram_range=(2, 3))
```

#### Identifikasi pola untuk subkategori Hair Dryer

```python
Hair_Dryers = df[df['Subkategori'] == 'Hair Dryers']['Judul produk']
```

```python
Hair_Dryers
```

```python
bersih = Hair_Dryers.apply(preprocess_text)

X = vectorizer.fit_transform(bersih)
sum_words = X.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

print(words_freq[:10])
```

Berdasarkan bigram dan trigram, kata yang paling banyak muncul untuk subkategori Hair Dryer adalah kalimat Hair Dryer

```python
df[df["Judul produk"].str.contains("Hair Dryer", case=False)]
```

ada 65 totalnya bersama yang aslinya Hair dryer, sedangkan asumsi awal tadi adalah hanya ada 14, maka sisanya akan dikategorikan sebagai Hair Dryers

```python
mask = df["Judul produk"].str.contains("Hair Dryer", case=False, na=False)
df.loc[mask, "Subkategori"] = df.loc[mask, "Subkategori"].fillna("Hair Dryers")
```
